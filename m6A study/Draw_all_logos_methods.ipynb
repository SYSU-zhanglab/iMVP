```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
```

```python
orders1 = ["CIMS", "CITS", "m6ACE-seq_all", "m6A-label-seq"]
orders2 = ["MAZTER-seq", "m6A-REF-seq", "DART-seq", "xPore"]
```

```python
fig, axes = plt.subplots(7, 4, figsize=(16,8))
for i, path in zip(*[range(4), orders1]):
    j = 0
    for x in [1,2,3,4,5,6,7]:
        img = plt.imread("./"+path+"/logos_bits_png_no_axis/cluster_{}.png".format(x))
        _ = axes[j][i].imshow(img)
        j += 1
for ax in axes.reshape(-1):
    ax.axis("off")
plt.tight_layout()
plt.subplots_adjust(left=0, right=1., top=1., bottom=0., hspace=0.0, wspace=0.05)
plt.savefig("m6A_methods_logos_1.pdf", dpi=1000)
```
