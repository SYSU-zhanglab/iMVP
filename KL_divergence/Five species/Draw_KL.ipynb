```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['font.family'] = 'arial'
mpl.rcParams['xtick.labelsize'] = 9
mpl.rcParams['ytick.labelsize'] = 9
```

```python
df = pd.read_csv("KL_species_5bases.csv")
```

```python
df2 = df.pivot(index="SpeciesA", columns="SpeciesB", values="KL")
df2 = df2.loc[["Human", "Mouse", "XT", "XL", "ZF", "Fly"]]
df2 = df2[["Human", "Mouse", "XT", "XL", "ZF", "Fly"]]
```

```python
mask = np.triu(np.ones_like(df2))
```

```python
fig, ax = plt.subplots()
sns.heatmap(df2, square=True, mask=mask, annot=True, cmap="Reds")
plt.savefig("KL_divergence.pdf")
```
